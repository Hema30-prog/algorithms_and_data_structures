## Question Three: Move All Zeros to End
**Description**  
Given a list of integers, move all zeros to the end while maintaining the relative order of the non-zero elements. 

**Examples**
```python
Input: nums = [0, 1, 0, 3, 12]
Output: [1, 3, 12, 0, 0]
```
```python
Input: nums = [4, 0, 5, 0, 0, 6]
Output: [4, 5, 6, 0, 0, 0]
```


````python
from typing import List

def move_zeros_to_end(nums: List[int]) -> List[int]:
    # TODO

    # Collect all non-zero elements
    result = [num for num in nums if num != 0]

    # Count the number of zeros
    zero_count = len(nums) - len(result)

    # Append zeros to the end
    result.extend([0] * zero_count)

    return result
  
````


-   Code the solution to your assigned problem in Python (code chunk). Note: each problem can be solved more simply if you use an abstract data type that is suitable for that problem. Using that try to find the best time and space complexity solution!


````python
# Your answer here
# "write pointer” approach -an efficient abstract data type idea: treat the list like a stream and overwrite in-place
from typing import List

def move_zeros_to_end(nums: List[int]) -> List[int]:
    write_index = 0

    # Move all non-zero elements forward
    for num in nums:
        if num != 0:
            nums[write_index] = num
            write_index += 1

    # Fill remaining positions with zeros
    for i in range(write_index, len(nums)):
        nums[i] = 0

    return nums
````

````python
nums = [0, 1, 0, 3, 12]
print(move_zeros_to_end(nums))
````

-   Explain the problem’s time and space complexity


# Your answer here
Time Complexity: O(n)

Write-pointer approach
First loop: visit each element once → O(n)
Second loop (fill zeros): worst case touches all elements → O(n)
So:  O(n) + O(n) = O(n)

Space Complexity: O(1)

We only use a few variables:

-write_index 
-num

We do not use extra arrays, queues, or lists

So Constant extra memory regardless of input size → O(1)
