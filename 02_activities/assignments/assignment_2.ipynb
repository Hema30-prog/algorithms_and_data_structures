  You are given a list containing `n` integers in the range `[1, n]`. Return a list of numbers that are missing from the range `[1, n]` of the array. If there is no missing number, return -1. Note, all the integers in the list may not be unique.
  
  ## Examples

  ### Example 1

  Input: `lst = [1, 1, 3]`

  Output: [2]

  ### Example 2

  Input: `lst = [5,3,2,4,1]`

  Output: [-1]

  ### Example 3

  Input: `lst = [6, 8, 2, 3, 5, 7, 1, 1, 2,10]`

  Output: [4, 9]

</details>

#### Starter Code for Question 3


```python
from typing import List

def missing_num(nums: List[int]) -> List[int]:
    n = len(nums)

    # Mark visited numbers
    for i in range(n):
        index = abs(nums[i]) - 1
        if 0 <= index < n and nums[index] > 0:
            nums[index] = -nums[index]

    missing = []

    # Positive values indicate missing numbers
    for i in range(n):
        if nums[i] > 0:
            missing.append(i + 1)

    return missing if missing else [-1]
```

Given a string made up only of the characters (, ), {, }, [ and ]. The task is to determine whether the brackets in the string are arranged correctly.

# Trace / Walkthrough of the Example that Partner made in assignment 1

Example:

print(is_valid_brackets("([]{})"))

Expected Output:

True
Step-by-Step Execution

The function starts by creating:

pairs = {')': '(', '}': '{', ']': '['}
stack = []

The stack keeps track of opening brackets that have not yet been matched.

| Character | Action | Stack State |
|----------|--------|-------------|
| `(` | Opening bracket → pushed to stack | `['(']` |
| `[` | Opening bracket → pushed to stack | `['(', '[']` |
| `]` | Matches top `'['` → pop from stack | `['(']` |
| `{` | Opening bracket → pushed to stack | `['(', '{']` |
| `}` | Matches top `'{'` → pop from stack | `['(']` |
| `)` | Matches top `'('` → pop from stack | `[]` |

Final Check

After every character has been processed, the stack is empty:

return len(stack) == 0

Since the stack contains no unmatched opening brackets, the function returns True.


-   Copy the solution your partner wrote. 


```python
def is_valid_brackets(s: str) -> bool:
    pairs = {')': '(', '}': '{', ']': '['}
    stack = []

    for char in s:
        if char in '({[':
            stack.append(char)
        elif char in ')}]':
            if not stack or stack.pop() != pairs[char]:
                return False
    
    return len(stack) == 0
```


-   Explain why their solution works in your own words.


This solution works because it uses a stack (Last-In, First-Out structure) to correctly match opening and closing brackets in the order they appear.

This approach guarantees two key rules:

Correct pairing → every closing bracket matches the right type
Correct order → the most recent opening bracket is matched first

Together, these conditions fully define a valid bracket sequence.


-   Explain the problem’s time and space complexity in your own words.


**Time Complexity (O(n))**
The algorithm goes through the string once, checking each character one by one.

Each bracket is processed only a single time.
Stack operations (push and pop) take constant time, O(1).

So even if the string gets longer, the work grows directly in proportion to the input size.

That’s why the time complexity is O(n).

**Space Complexity (O(n))**
The extra memory used comes from the stack.

In the worst case, all characters could be opening brackets like "((({{{[[["
In that situation, nothing gets removed, so the stack grows to store all n characters.

So the memory needed increases with the size of the input.

That’s why the space complexity is O(n).
